```python
#homework02
import csv

students = []
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    for row in reader:
        name, age, grade = row[0], int(row[1]), float(row[2])
        students.append((name, age, grade))

high_grade_students = [s for s in students if s[2] >= 7]

for s in high_grade_students:
    print(s)

with open("high_grade_students.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    for student in high_grade_students:
        writer.writerow(student)
```

```python
#homework03
import json

with open("user_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

for user in data:
    if user.get("id") == 3:
        print("Tên:", user.get("name"))
        print("Tuổi:", user.get("age"))
        break

new_id = int(input("Nhập id mới: "))
new_name = input("Nhập tên mới: ")
new_age = int(input("Nhập tuổi mới: "))

data.append({
    "id": new_id,
    "name": new_name,
    "age": new_age
})

with open("user_data.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
```

```python
#homework06
import json

with open("inventory.json", "r", encoding="utf-8") as f:
    data = json.load(f)

new_id = input("Nhập product_id: ")
new_name = input("Nhập tên sản phẩm: ")
new_quantity = int(input("Nhập số lượng: "))
new_price = float(input("Nhập giá: "))


new_product = {
    "product_id": new_id,
    "name": new_name,
    "quantity": new_quantity,
    "price": new_price
}

data.append(new_product)


with open("inventory.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
```

```python
#homework08
from datetime import datetime

today = datetime.today().date()

with open("log.txt", "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue

        #format: "2025-11-23 14:55:01 - message"
        timestamp_str = line.split(" - ")[0]

        try:
            ts = datetime.strptime(timestamp_str, "%Y-%m-%d %H:%M:%S")
            if ts.date() == today:
                print(line)
        except ValueError:
            continue
```
